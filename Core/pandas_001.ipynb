# Filter Rows, View Selected Rows

## Using iloc() method

```python
df.iloc[0]
```

```python
df.iloc[0][2]
```

```python
df.iloc[[0,1]]
```

```python
df.iloc[[0,1]]['sex']
```

## Using loc() method

```python
df.loc[df['sex']=='male']
```

```python
df.loc[(df['sex']=='male') & (df['fare']>200)]
```

## Also we can apply the above condition without using loc() function (Boolean Indexing)

```python
df[(df['sex']=='male') & (df['fare']>200)]
```

## Selecting both Rows and COlumns

```python
df[(df['sex']=='male') & (df['fare']>200)][['survived','pclass', 'sex', 'fare']]
```

## Using isin operator

```python
df.loc[df['class'].isin(['First','Second'])]
```
